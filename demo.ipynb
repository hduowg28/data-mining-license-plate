```python
def write_file(file_name="data_test.txt", data):
    with open(file_name, 'w', encoding='utf-8') as file:
        for row in data:
            file.write(f"{row[0]},{row[1]},{row[2]},{row[3]}\n")
    print("luu file thanh cong\n")
```

```python
from ultralytics import YOLO

model = YOLO("C:/data mining/best.pt")
results=model.predict(source="dataset_license_plate/test2")
for result in results:
    for box in result.boxes:
        data = map(int, box.xyxy[0])
        write_file(data=data)
        
```
